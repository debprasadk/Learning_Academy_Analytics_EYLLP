```python
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
taxi = sns.load_dataset('taxis')
taxi.head()
```

```python
sns.barplot(x='payment',y='fare',data=taxi);
```

```python
sns.barplot(x='payment',y='fare',data=taxi,estimator=np.sum)
```

```python
taxi['payment'].value_counts()
```

```python
sns.countplot(x='payment',data=taxi)
```

```python
sns.boxplot(x='payment',y='fare',data=taxi, palette='rainbow')
```

```python
sns.boxplot(x="color", y="fare", data=taxi, palette='rainbow')
```

```python
taxi.corr()
```

```python
sns.heatmap(taxi.corr())
```

```python
sns.heatmap(taxi.corr(),cmap='coolwarm',annot=True)
```
